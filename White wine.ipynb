```python
# Import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('https://assets.datacamp.com/production/repositories/628/datasets/2d9076606fb074c66420a36e06d7c7bc605459d4/white-wine.csv')
```

```python
# Plot out pairplots for 4 columns at a time with 'quality' as the target
excol = ['free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'sulphates', 'alcohol']
df1 = df.drop(excol, axis=1)
sns.set_theme()
sns.pairplot(df1, hue='quality')
```

```python
# Plot out pairplots for 4 columns at a time with 'quality' as the target
excol = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides']
df2 = df.drop(excol, axis=1)
sns.pairplot(df2, hue='quality')
```

```python
# Review correlations to 'quality'
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(df.corr(), square=True, cmap='RdYlGn')
```
