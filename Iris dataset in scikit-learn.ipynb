# Use scikit-learn to fit a classifier

```python
# import the K Nearest Neighbors classifier and instantiate it
from sklearn.neighbors import KNeighborsClassifier 
```

knn requires:
* training data is either a numpy array or pandas series
* the feature data takes on continuous values NOT categories
* no missing values in the data

```python
# Looking at the data using shape indicates we have 150 rows (observations) and 4 columns (features) in the data
iris['data'].shape
```

```python
# Looking at the labels using shape indicates we also have 150 rows (observations) and 1 column (labels) in the target
iris['target'].shape
```

```python
# Instantiate the knn classifier with n_neighbors set to 6
knn = KNeighborsClassifier(n_neighbors=6)
```

```python
# Fit the classifier to the training data
knn.fit(iris['data'], iris['target'])
```

# Predicting on unlabeled data

```python
# Use data the classifier has not seen previously
X_new = np.array([[5.6,2.8,3.9,1.1],
                [5.7,2.6,3.8,1.3],
                [4.7,6.2,1.3,0.2]])
```

```python
X_new.shape
```

The X_new array has 3 rows and 4 features so we would expect the .predict() method to return a [1,3] array

```python
# Use the fit model to predict
prediction = knn.predict(X_new)
print(f"Prediction: {prediction}")
```

## Train/test split

In order to measure the 'goodness' of the predictions we need a metric. Typically we don't want to run this metric on the same data we used to train the model, so we need to have "hold out" data that has labels that the model has never seen. This will give us an idea of how well the model with "generalize" to new data which it has never seen before. 

We can set aside a portion of our original labeled data and call it "test" and then use the remaining data to "train" the model. Then we use the .predict() method on the test data. Since we know the labels for this test data we can then compare how well the model did in predicting the known labels.

One measure used is the Accuracy defined as:
The fraction of correct predictions (# of correct predictions / # of test data points)

```python
# Import dependencies
from sklearn.model_selection import train_test_split
```

```python
# create the train/test arrays for both the training  data and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                  random_state=21, stratify=y)
```

```python
# Instantiate our knn model
knn = KNeighborsClassifier(n_neighbors=8)
```

```python
# Fit the model
knn.fit(X_train, y_train)
```

```python
# Predict using the X_test data
y_pred = knn.predict(X_test)
print(f"Test set predictions:\n {y_pred}")
```

```python
# Score the Accuracy of the model from the test set predictions
knn.score(X_test, y_test)
```

# Confusion matrix

```python
# Import dependencies
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

# Overfitting and underfitting

Using different values of n_neighbors will result in different accuracy scores for both the training and test data.
To visualize the impact this has we can graph it out

```python
# Setup arrays to store train and test accuracies
neighbors = np.arange(1, 9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
```

```python
# Loop over different values of k
for i, k in enumerate(neighbors):
    # Setup a k-NN Classifier with k neighbors: knn
    knn = KNeighborsClassifier(n_neighbors=k)

    # Fit the classifier to the training data
    knn.fit(X_train, y_train)
    
    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(X_train, y_train)

    #Compute accuracy on the testing set
    test_accuracy[i] = knn.score(X_test, y_test)
```

```python
# Generate plot
fig, ax = plt.subplots(figsize=(15,10))
plt.title('k-NN: Varying Number of Neighbors')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

# Unsupervised Learning on Iris dataset

```python
# Import dependencies
from sklearn.cluster import KMeans
```

```python
# Instantiate the model
model = KMeans(n_clusters=3)
```

```python
# Fit the model
model.fit(X)
```

```python
# Predict from the model and generate labels for clusters
labels = model.predict(X)
print(labels)
```

```python
# Visualize by scatter plot the clusters using sepal length (zeroth column) and petal length (second column)
xs = X[:,0]
ys = X[:,2]
fig, ax = plt.subplots(figsize=(15,10))
plt.scatter(xs, ys, c=labels)
plt.xlabel("Sepal length")
plt.ylabel("Petal length")
```

## Use t-SNE clustering

## Dimension reduction using PCA

```python
# Import dependencies
from sklearn.decomposition import PCA
```

```python
# Instantiate the PCA model with n_components set to 2
pca = PCA(n_components=2)
```

```python
# Fit and transform the samples
pca.fit(X)
transformed = pca.transform(X)
print(transformed.shape)
```

```python
# Plot the pca features in a scatter plot
fig, ax = plt.subplots(figsize=(15,10))
xs = transformed[:,0]
ys = transformed[:,1]
plt.scatter(xs, ys, c=labels, edgecolors='b')
```
